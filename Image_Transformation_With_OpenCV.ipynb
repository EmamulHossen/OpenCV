# **Importing Necessary Modules**

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

**Read Image**

```python
image=cv.imread("/content/drive/MyDrive/Deep Learning-12/My_Code/Cat_image_Transformation/shutterstock_56602966-2.jpg")
```

**Define a user define feunction, the function name is image_display**

```python
def image_display(imageTitle,image):
   if len(image.shape)==3:
     imageVariable=cv.cvtColor(image,cv.COLOR_RGB2BGR)

   plt.figure(figsize=(5,5))
   plt.imshow(imageVariable)
   plt.title(imageTitle)
   plt.show()
```

```python
image_display("This is Cat Image", image)
```

# **TRASLATION**
>A translation moves an object to a different position on the screen. You can translate a point in 2D by adding translation coordinate (tx, ty) to the original coordinate X,Y
 to get the new coordinate X′,Y′


 translation.png
.



```python
translationx=300 #Pixel Value
translationy=150 #Pixel Value

modifiedImage=np.float32([[1,0,translationx],
                          [0,1,translationy]])

modifiedImage
```

```python
shifted_image=cv.warpAffine(image,modifiedImage,(image.shape[1],(image.shape[0])))
```

```python
image_display("This is Cat Image", image)
image_display("Shifted Image",shifted_image)
```

# **ROTATION**
> In rotation, we rotate the object at particular angle θ theta
 from its origin. From the following figure, we can see that the point PX,Y
 is located at angle φ from the horizontal X coordinate with distance r from the origin.

> Let us suppose you want to rotate it at the angle θ. After rotating it to a new location, you will get a new point P’ X′,Y′


Rotation.png

.



```python
imageCenter= (image.shape[1]//2,image.shape[0]//2)
imageRotate=cv.getRotationMatrix2D(imageCenter,30,1)
finalRotate=cv.warpAffine(image,imageRotate,(image.shape[1],(image.shape[0])))
```

```python
image_display("This is Cat Image", image)
image_display("Rotate Image",finalRotate)
```

# **IMAGE RESIZE**
> Image resizing in OpenCV is important for a variety of applications, such as preparing images for analysis, reducing computational complexity, changing image dimensions for display, or fitting images to a particular format. When resizing an image, OpenCV employs interpolation techniques to estimate the pixel values of the new image. Common interpolation methods include nearest-neighbor, bilinear, and cubic interpolation.



```python
def image_Resize (image,imagesize):
  imageResize=cv.resize(image,(imagesize,imagesize),interpolation=cv.INTER_AREA)
  print("Before resize image shape is : {}".format(image.shape))
  image_display("This is Cat Image", image)
  print("After  resize image shape {}".format(imageResize.shape))
  image_display("Resize Image",imageResize)
```

```python
image_Resize(image,300)
```

# **IMAGE FLIPPING**

```python
def image_flipping (image):
  imageflipping=cv.flip(image,0)
  print("Before flipping image shape is : {}".format(image.shape))
  image_display("This is Cat Image", image)
  print("After flipping  image shape {}".format(imageflipping.shape))
  image_display("Flipping Image",imageflipping)
```

```python
image_flipping(image)
```
