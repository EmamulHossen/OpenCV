# **Edge Detection Using Gradient Method**

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt


# Load an image
image = cv2.imread('/content/drive/MyDrive/Deep Learning-12/Class-7/Cat (1).jpg', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian smoothing to reduce noise
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Apply Sobel operators to calculate gradients
grad_x = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
grad_y = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)

# Compute the gradient magnitude and direction
magnitude = np.sqrt(grad_x**2 + grad_y**2)
direction = np.arctan2(grad_y, grad_x)

# Define a threshold to detect edges
threshold = 10

# Create a binary edge map
edge_map = np.zeros_like(image)
edge_map[magnitude > threshold] = 255  # Set edge pixels to white

# Display the original image and the edge map
plt.figure(figsize=(5,5))

```

```python
plt.imshow( image)
plt.title("Orginal Image")
plt.show()
```

```python
plt.imshow( edge_map)
plt.title("Edge Image")
plt.show()
```

# **Edge Detection Using Laplacian**

```python
import cv2

# Load an image
image = cv2.imread('/content/drive/MyDrive/Deep Learning-12/Class-7/Cat.png', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian smoothing to reduce noise
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Apply Laplacian filter
laplacian = cv2.Laplacian(blurred, cv2.CV_64F)

# Calculate the absolute value of the Laplacian
laplacian_abs = cv2.convertScaleAbs(laplacian)

# Define a threshold to detect edges
threshold = 10

# Create a binary edge map
edge_map = cv2.threshold(laplacian_abs, threshold, 255, cv2.THRESH_BINARY)[1]

# Display the original image and the edge map
plt.figure(figsize=(5,5))

```

```python
plt.imshow( image)
plt.title("Orginal Image")
plt.show()
```

```python
plt.imshow( edge_map)
plt.title("Edge Image")
plt.show()
```
